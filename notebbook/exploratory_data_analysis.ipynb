```python
# Convert dates
date_cols = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'next_pymnt_d',
             'last_credit_pull_d', 'hardship_start_date', 'hardship_end_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Create year & quarter
df['year'] = df['issue_d'].dt.year  
df['quarter'] = df['issue_d'].dt.to_period('Q')
```

```python
# Missing values analysis
missing = df.isnull().mean().sort_values(ascending=False)
print("Top Missing Values:")
print(missing.head(20))
 
```

```python
# Target Variable Engineering
def label_default(x):
    if isinstance(x, str) and (
        'Charged Off' in x or 'Default' in x or 
        'Does not meet the credit policy. Status:Charged Off' in x
    ):
        return 1
    else:
        return 0

df['is_default'] = df['loan_status'].apply(label_default)
print(df['is_default'].value_counts(normalize=True))
```

```python
# Univariate Analysis
num_cols = ['loan_amnt', 'int_rate', 'annual_inc', 'dti', 'fico_range_low', 'installment']
df[num_cols].describe().T
```

```python
df[num_cols].hist(bins=30, figsize=(12,8))
plt.suptitle('Distribution of Key Numeric Features')
plt.show()
```

```python
cat_cols = ['term', 'grade', 'home_ownership', 'verification_status', 'purpose']
for col in cat_cols:
    plt.figure(figsize=(6,3))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
# Clean numeric percentage columns
if df['int_rate'].dtype == 'object':
    df['int_rate'] = df['int_rate'].str.replace('%', '', regex=False).astype(float)

if 'revol_util' in df.columns and df['revol_util'].dtype == 'object':
    df['revol_util'] = df['revol_util'].str.replace('%', '', regex=False).astype(float)
```

```python
# Bivariate Analysis - Numeric vs Target
plt.figure(figsize=(6,4))
sns.boxplot(x='is_default', y='int_rate', data=df)
plt.title("Interest Rate vs Default")

plt.figure(figsize=(6,4))
sns.boxplot(x='is_default', y='annual_inc', data=df)
plt.title("Annual Income vs Default")

plt.figure(figsize=(6,4))
sns.barplot(x='grade', y='is_default', data=df, estimator=np.mean)
plt.title("Default Rate by Grade")

plt.figure(figsize=(6,4))
sns.barplot(x='home_ownership', y='is_default', data=df, estimator=np.mean)
plt.title("Default Rate by Home Ownership")
```

```python
# Correlation Heatmap
corr = df[num_cols + ['is_default']].corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Temporal Trends
defaults_by_year = df.groupby('year')['is_default'].mean()
plt.figure(figsize=(8,4))
defaults_by_year.plot(marker='o')
plt.title("Default Rate Over Years (2008–2020Q1)")
plt.ylabel("Default Rate")
plt.show()
```

```python
loan_volume = df.groupby('year')['loan_amnt'].count()
loan_volume.plot(kind='bar', color='skyblue', title="Number of Loans Issued per Year")
plt.show()
```

```python
# Regional Analysis
state_defaults = df.groupby('addr_state')['is_default'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,5))
state_defaults.plot(kind='bar', color='orange')
plt.title("Default Rate by State")
plt.ylabel("Default Rate")
plt.show()
```

```python
path = r'..\data\unemployment_rate_by_state.csv'
fred_df = pd.read_csv(path)
fred_df['DATE'] = pd.to_datetime(fred_df['DATE'])
fred_df['year'] = fred_df['DATE'].dt.year
fred_df['quarter_num'] = ((fred_df['DATE'].dt.month - 1) // 3 + 1)
fred_df['quarter'] = fred_df['year'].astype(str) + 'Q' + fred_df['quarter_num'].astype(str)

fred_q = fred_df.groupby(['addr_state', 'quarter'], as_index=False)['unemployment_rate'].mean()


df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y', errors='coerce')
df['year'] = df['issue_d'].dt.year
df['quarter_num'] = ((df['issue_d'].dt.month - 1) // 3 + 1)
df['quarter'] = df['year'].astype(str) + 'Q' + df['quarter_num'].astype(str)
df['quarter'] = df['quarter'].astype(str)
df['quarter'] = df['quarter'].str.replace('.0', '', regex=False)

merged_df = pd.merge(df, fred_q, how='left', on=['addr_state', 'quarter'])

merged_df['is_default'] = merged_df['loan_status'].apply(lambda x: 1 if 'Charged Off' in str(x) else 0)

sns.scatterplot( y='is_default',x='unemployment_rate', data=merged_df, alpha=0.3)
plt.title("Unemployment Rate vs Default Probability")
plt.xlabel("State Unemployment Rate")
plt.ylabel("Default (1 = Yes)")
plt.show()
avg_default = merged_df.groupby('unemployment_rate')['is_default'].mean().reset_index()

sns.lineplot(x='unemployment_rate', y='is_default', data=avg_default, marker='o')
plt.title("Average Default Rate vs State Unemployment Rate")
plt.xlabel("State Unemployment Rate (%)")
plt.ylabel("Average Default Probability")
plt.show()
```

```python
#1. Convert date & extract year/quarter
# Clean issue date
df['issue_d'] = pd.to_datetime(df['issue_d'], errors='coerce')

# Extract year and quarter
df['issue_year'] = df['issue_d'].dt.year
df['issue_quarter'] = df['issue_d'].dt.to_period('Q')
```

```python
# Loan Count & Loan Amount Trend (Yearly)
yearly_loans = df.groupby('issue_year')['id'].count()
yearly_amount = df.groupby('issue_year')['loan_amnt'].sum()

plt.figure(figsize=(12,5))
sns.lineplot(x=yearly_loans.index, y=yearly_loans.values, marker='o')
plt.title("Number of Loans Issued Per Year")
plt.xlabel("Year")
plt.ylabel("Loan Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,5))
sns.lineplot(x=yearly_amount.index, y=yearly_amount.values, marker='o')
plt.title("Total Loan Amount Issued Per Year")
plt.xlabel("Year")
plt.ylabel("Total Loan Amount ($)")
plt.grid(True)
plt.show()
```

```python
##  Simplify Loan Status for Default Analysis
df['loan_status_clean'] = df['loan_status'].replace({
    'Fully Paid': 'Paid',
    'Charged Off': 'Defaulted'
})
```

```python
#  Default Rate by Year (Stacked Bar Chart)
status_year = (df.groupby(['issue_year', 'loan_status_clean'])
                 .size()
                 .unstack()
                 .fillna(0))

status_year_pct = status_year.div(status_year.sum(axis=1), axis=0) * 100

status_year_pct[['Paid', 'Defaulted']].plot(kind='bar', stacked=True, figsize=(14,6))
plt.title("Loan Status Distribution by Year")
plt.xlabel("Issue Year")
plt.ylabel("Percentage (%)")
plt.legend()
plt.show()
```

```python
# Default Rate Trend Over Time (Line Chart)
default_rate = (
    df.groupby('issue_year')
      .apply(lambda x: (x['loan_status_clean'] == 'Defaulted').mean() * 100)
      .reset_index(name='default_rate')
)

plt.figure(figsize=(12,5))
sns.lineplot(data=default_rate, x='issue_year', y='default_rate', marker='o')
plt.title("Default Rate Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Default Rate (%)")
plt.grid(True)
plt.show()
```

```python
# Default Rate Trend Over Time (Line Chart)
quarterly = (
    df.groupby(['issue_quarter', 'loan_status_clean'])
      .size()
      .reset_index(name='count')
)

quarterly['total'] = quarterly.groupby('issue_quarter')['count'].transform('sum')
quarterly['perc'] = quarterly['count'] / quarterly['total'] * 100

#  Convert Period to string for plotting
quarterly['issue_quarter'] = quarterly['issue_quarter'].astype(str)

plt.figure(figsize=(15,6))
sns.lineplot(data=quarterly, x='issue_quarter', y='perc', hue='loan_status_clean', marker='o')
plt.title("Loan Status % by Quarter")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Grade Distribution Over Time (Count Plot)
# Clean interest rate column — remove '%' and convert to float
df['int_rate'] = (
    df['int_rate']
    .astype(str)
    .str.replace('%', '', regex=False)
    .str.strip()
    .astype(float)
)

# Now safely group by year
interest_trend = df.groupby('issue_year')['int_rate'].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=interest_trend, x='issue_year', y='int_rate', marker='o', color='purple')
plt.title("Average Interest Rate Trend Over Time")
plt.xlabel("Issue Year")
plt.ylabel("Average Interest Rate (%)")
plt.grid(True)
plt.show()
```

```python
# Grade Distribution Over Time (Count Plot)
plt.figure(figsize=(14,6))
sns.countplot(data=df, x='issue_year', hue='grade')
plt.title("Grade Distribution Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
#  Unemployment Rate vs Default Rate
# Already loaded fred_df in your file
fred_year = fred_df.groupby('year')['unemployment_rate'].mean().reset_index()

merged = default_rate.merge(fred_year, left_on='issue_year', right_on='year')

plt.figure(figsize=(12,5))
sns.scatterplot(data=merged, x='unemployment_rate', y='default_rate')
plt.title("Unemployment Rate vs Default Rate")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Default Rate (%)")
plt.grid(True)
plt.show()
```
